```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import constants as const
mass, distance=np.genfromtxt('ExoplanetData.csv', delimiter=',', skip_header=2, usecols=(0,1), filling_values=0, unpack=True)
def FitFunc(x, a, b):
    return a * np.sin(b * x)
params, params_cov = curve_fit(FitFunc, mass, distance, p0=[2, 2])
plt.clf()
plt.loglog(mass, distance, 'ob', markersize=3, alpha=0.3, label='Exoplanet.org Data')
#plt.plot(mass, FitFunc(mass, params[0], params[1]), '--', linewidth=1, color='orange', label='Fitted Data')
plt.title('Exoplanet Mass vs. Stellar Distance', size=15)
plt.xlabel('Mass [M(J)]', size=14)
plt.ylabel('Distance to Star [AU]', size=14)
plt.xticks(size=13)
plt.yticks(size=13)
plt.legend(edgecolor='black')
plt.show()
```
